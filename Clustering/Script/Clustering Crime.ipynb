```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python

# Load data sets
crime = pd.read_csv("F:\Data Science\Assignment\Clustering\Data\crime_data.csv")
```

```python
crime.head()
```

```python
crime.drop(['Unnamed: 0'],axis=1,inplace=True)
crime.head()
```

```python
crime.info() 
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Normalized data frame (considering the numerical part of data)
crime_norm = norm_func(crime)
crime_norm
```

```python
# Create Dendrograms
plt.figure(figsize=(10, 7))  
dendograms=sch.dendrogram(sch.linkage(crime_norm,'complete'))
```

```python
# Create Clusters (y)
hclusters=AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='complete')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(crime_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
# Adding clusters to dataset
crime['clustersid']=hclusters.labels_
crime
```

```python
crime.groupby('clustersid').agg(['mean']).reset_index()

```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(crime['clustersid'],crime['Murder'], c=hclusters.labels_) 

```

# K means clustering

```python
# Import Libraries
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
# Load data sets
crime = pd.read_csv("F:\Data Science\Assignment\Clustering\Data\crime_data.csv")
```

```python
crime.drop(['Unnamed: 0'],axis=1,inplace=True)
crime.head()
```

```python
# Normalization function 
def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
# Normalized data frame (considering the numerical part of data)
crime_norm = norm_func(crime)
```

```python
crime_norm
```

```python
# Selecting 4 clusters from the above scree plot which is the optimum number of clusters 
model = KMeans(n_clusters=4)
model.fit(crime_norm)
model.labels_                    # getting the labels of clusters assigned to each row 
```

```python
md = pd.Series(model.labels_)  # converting numpy array into pandas series object 
crime['clusters'] = md          # creating a  new column and assigning it to new column 
crime
```

```python
crime.groupby(crime.clusters).mean()
```

```python
# Plot Clusters
plt.figure(figsize=(10, 7))  
plt.scatter(crime['clusters'],crime['Murder'], c=model.labels_)

```
